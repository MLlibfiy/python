```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({"x": [1, 2, 3, 4, 5, 6, 7], "y": [1, 2, 3, 4, 5, 6, 7]})
print(df)

sns.set(style="white")
sns.jointplot(x="x",y="y",data=df)
```

```python
data = pd.read_csv("data/loandata.csv")
sns.jointplot(x="total_pymnt",y="total_rec_int",data=data,kind="reg")
```

```python
sns.jointplot(x="total_pymnt",y="total_rec_int",data=data,kind="hex")
```

```python
sns.jointplot(x="total_pymnt",y="total_rec_int",data=data,space=0,color="b",kind="kde")
```
